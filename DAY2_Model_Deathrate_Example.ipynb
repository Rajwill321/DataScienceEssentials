```python
import pandas as pd
df = pd.read_csv(r"https://raw.githubusercontent.com/vishurudratrainer/notebooks/main/deathrate.csv")
```

```python
df.head()
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['A9'], df['A16'], color='blue')
plt.xlabel('A9')  
plt.ylabel('A16')
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(df[['A9', 'A6']], df['A16'], random_state=0)
```

```python
from sklearn.linear_model import LinearRegression  
lm = LinearRegression()  
lm.fit(X_train,y_train)   # model is generated here
```

```python
y_pred = lm.predict(X_test) # test the model
```

```python
result = pd.DataFrame({'actual': y_test, 'predicted': y_pred})
result.head(10)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test, y_pred))
print("R2: {:.2f}".format(r2_score(y_test, y_pred)))
print(f"R2: {r2_score(y_test, y_pred)}")
```
