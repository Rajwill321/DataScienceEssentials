```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips['tips_perc'] = tips['tip']/(tips['total_bill']-tips['tip'])
```

```python
sns.barplot(x='day', y='tips_perc', data=tips, orient='v')
```

```python
sns.barplot(x='day', y='tips_perc', hue='time', data=tips, orient='v')
```

```python
# 3rd dimention is size now, and we see tip highest on saturday
sns.barplot(x='day', y='tips_perc', hue='size', data=tips, orient='v')
```

```python
# regulation plot will approximate the value
sns.regplot(x='sepal.length', y='petal.length', data=iris)
```
