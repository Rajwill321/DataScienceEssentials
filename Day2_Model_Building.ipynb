```python
import pandas as pd
import numpy as np
from sklearn import datasets
boston_df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
#here we are directly renaming the medv to target
boston_df.rename(columns={"medv": "target"}, inplace=True)
boston_df.head()
```

```python
boston_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = boston_df.corr().round(2)
plt.figure(figsize=(14, 12))  
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
plt.scatter(boston_df['lstat'], boston_df['target'], color='black')
plt.xlabel('LSTAT')  
plt.ylabel('price')
```

```python
plt.scatter(boston_df['rm'], boston_df['target'], color='black')
plt.xlabel('RM')  
plt.ylabel('price')
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(boston_df[['lstat']], boston_df['target'], random_state=0)
```

```python
from sklearn.linear_model import LinearRegression  
lm = LinearRegression()  
lm.fit(X_train,y_train)   # model is generated here
```

```python
y_pred = lm.predict(X_test) # test the model
```

```python
y_pred
```

```python
result = pd.DataFrame({'actual': y_test, 'predicted': y_pred})
result.head(10)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test, y_pred))
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(boston_df[['lstat', 'rm']], boston_df['target'], random_state=0)
```

```python
from sklearn.linear_model import LinearRegression  
lm = LinearRegression()  
lm.fit(X_train,y_train)   # model is generated here
```

```python
y_pred = lm.predict(X_test) # test the model
```

```python
result = pd.DataFrame({'actual': y_test, 'predicted': y_pred})
result.head(10)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_test, y_pred))
print(f"R2: {r2_score(y_test, y_pred)}")
```
