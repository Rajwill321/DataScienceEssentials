```python
import pandas as pd
```

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
r1=pd.DataFrame(raw_data_1)
```

```python
r1
```

```python
r2=pd.DataFrame(raw_data_2)
r3=pd.DataFrame(raw_data_3)
```

```python
r3
```

```python
# Inner Join 
r1_r3 = pd.merge(r1, r3, left_on='subject_id', right_on='subject_id') 
```

```python
r1_r3
```

```python
# Left outer join
r1_r3 = pd.merge(r1, r3, left_on='subject_id', right_on='subject_id', how='left') 
```

```python
r1_r3
```

```python
# Right join
r1_r3 = pd.merge(r1, r3, left_on='subject_id', right_on='subject_id', how='right')
```

```python
r1_r3
```

```python
 # Outer join
 r1_r3  = pd.merge(r1, r3, left_on='subject_id', right_on='subject_id', how='outer')
```

```python
r1_r3
```

```python
r1_r3 = pd.merge(r1, r3, how='cross')
```

```python
r1_r3
```

```python
r1_r3 = pd.merge(r1, r3, left_on=['subject_id'],right_on=['subject_id'], how='outer')
```
