```python
import pandas as pd
```

```python
bank = pd.read_csv("c:\\Data\\bank-full.csv", sep=";")
bank.head()
```

```python
bank.dtypes
```

```python
col = [ 'job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome' ]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for c in col:
    bank[c] = le.fit_transform(bank[c])
    
bank.head()    
```

```python
bank.dtypes
```

```python
X=bank.drop("y", axis=1)
Y=bank['y']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.7, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logregmodel = LogisticRegression(random_state=0)
logregmodel.fit(X_train, Y_train)
```

```python
Y_pred = logregmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```
