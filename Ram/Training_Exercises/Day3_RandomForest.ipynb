```python
import pandas as pd
```

```python
wind=pd.read_csv("https://raw.githubusercontent.com/nikhilpodila/Classification-UCI-Datasets/master/winequality-red.csv", sep=";")
```

```python
wind.head()
```

```python
X=wind.drop('quality', axis=1)
Y=wind['quality']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
result = pd.DataFrame({"Actual": Y_test, "Predicted": Y_pred})
result
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```

```python
inds = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/ionosphere.csv", header=None)
```

```python
inds.head()
```

```python
col = []
for i in range(0,35):
    col.append("IP" + str(i))
print(col)
```

```python
inds.columns=col
inds.head()
```

```python
X=inds.drop('IP34',axis=1)
Y=inds['IP34']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)
result = pd.DataFrame({"Actual": Y_test, "Predicted": Y_pred})
result
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(Y_test, Y_pred))
print(accuracy_score(Y_test, Y_pred))
```
