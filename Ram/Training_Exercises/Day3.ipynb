```python
import pandas as pd
```

```python
iris=pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```python
iris.head()
```

```python
X=iris.drop('variety',axis=1)
```

```python
Y=iris['variety']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
result=pd.DataFrame({"Actual": Y_test, "Predicted": Y_pred})
```

```python
result.head()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(Y_test, Y_pred))
```

```python
print(accuracy_score(Y_test, Y_pred))
```

```python
iris.shape
```

```python
from sklearn.tree import export_text
```

```python
print(export_text(decision_tree=model, feature_names=model.feature_names_in_.tolist()))
```

```python
boston_df=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
```

```python
boston_df
```

```python
boston_df.rename(columns={"medv": "target"}, inplace=True)
```

```python
boston_df.head()
```

```python
X=boston_df.drop('target', axis=1)
```

```python
Y=boston_df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
result = pd.DataFrame({"Actual": Y_test, "Predicted": Y_pred})
```

```python
result.head()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("MSE: {:.2f}".format(mean_squared_error(Y_test, Y_pred)))  
print("R2: {:.2f}".format(r2_score(Y_test, Y_pred)))
```

```python
from sklearn.tree import export_text
print(export_text(decision_tree=model, feature_names=model.feature_names_in_.tolist()))
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
result=pd.DataFrame({"Actual": Y_test, "Predicted": Y_pred})
result.head()
```
