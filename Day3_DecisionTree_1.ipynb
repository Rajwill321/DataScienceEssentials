```python
# https://raw.githubusercontent.com/jbrownlee/Datasets/master/abalone.csv
```

```python
abalone = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/abalone.csv", header=None)
```

```python
abalone.head()
```

```python
abalone.columns=['IP1', 'IP2', 'IP3', 'IP4', 'IP5', 'IP6', 'IP7', 'IP8', 'OP']
```

```python
abalone.head()
```

```python
abalone['OP'].unique()
```

```python
abalone['OP'].value_counts()
```

```python
X = abalone.drop('OP', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X["IP1"] = le.fit_transform(X["IP1"]) #convert from non-numeric to numeric
```

```python
X.head()
```

```python
Y = abalone['OP']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
result = pd.DataFrame({"Actual":Y_test, "Predicted":Y_pred})
```

```python
print(f"Accuracy_Score: {accuracy_score(Y_test, Y_pred)}")
```

```python
print(f"Confusion_Matrix: \n {confusion_matrix(Y_test, Y_pred)}")
```
