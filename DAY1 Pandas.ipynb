```python
#create a new column new_variety based on condition

iris.loc[iris['variety']=='Setosa', 'new_variety']='s'
iris.loc[iris['variety']=='Versicolor', 'new_variety']='ve'
iris.loc[iris['variety']=='Virginica', 'new_variety']='vi'
```
